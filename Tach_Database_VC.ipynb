```

# Tach lay combinationlist
combination_list = []
top_inCombination_duplicated_List= []
#down_inCombination_duplicated_List= []
top_inCombinationList = []
#down_inCombinationList = []
full_idlist=[]
loopArray = [100]

with open('drive/My Drive/TLCN 2/FashionVC/dataset/combinationlist.dat','r') as f:
    content = f.readlines()
    i=0
for line in content:
    if i == 100:
       break
    line = line[:-1].split('\t')
    line = line[0]+"\t"+line[1]
    combination_list.append(line)

    line = line[:-1].split('\t')
    top_inCombination_duplicated_List.append(line[0])
    #down_inCombination_duplicated_List.append(line[1])
    i= i+1
np.savetxt("drive/My Drive/TLCN 2/FashionVC/ds224/combinationlist.dat", np.array(combination_list), fmt="%s")


for i in top_inCombination_duplicated_List:
    if i not in top_inCombinationList:
        top_inCombinationList.append(i)


#print(top_inCombinationList)

# down_inCombinationList = []
# for i in down_inCombination_duplicated_List:
#     if i not in down_inCombinationList:
#         down_inCombinationList.append(i)
# print(down_inCombinationList)
# full_idlist = top_inCombinationList+ down_inCombinationList

# ###########################################################################
```

```
# lấy danh sách devdata  for top
devdata_toplist = []
with open('drive/My Drive/TLCN 2/FashionVC/dataset/devdata_tops.dat','r') as f:
    content = f.readlines()
    
for top in top_inCombinationList :
    for line in content :
        line = line[:-1].split('\t')
        if top == line[0] :
            line=  line[0]+"\t"+line[1]+"\t"+line[2]
            devdata_toplist.append(line)
np.savetxt("drive/My Drive/TLCN 2/FashionVC/ds224/devdata_tops.dat", np.array(devdata_toplist), fmt="%s")
```

```


# #############################################################################

# # lấy danh sách for traindata
traindata_list = []
with open('drive/My Drive/TLCN 2/FashionVC/dataset/traindata.dat','r') as f:
    content = f.readlines()
    
for line in content :
    line = line[:-1].split('\t')
    for combination_line in combination_list :
        combination_line = combination_line[:-1].split('\t')
        if (combination_line[0] == line[0] or combination_line[1] == line [1]):
            line=  line[0]+"\t"+line[1]
            traindata_list.append(line)
    
np.savetxt("drive/My Drive/TLCN 2/FashionVC/ds224/traindata.dat", np.array(traindata_list), fmt="%s")


# #############################################################################
```

```

# # lấy danh sách for toplist
toplist = []
with open('drive/My Drive/TLCN 2/FashionVC/dataset/toplist.dat','r') as f:
    content = f.readlines()

for line1 in traindata_list:
  line1 = line1[0:].split('\t')
  toplist.append(line1[0])
    
for line in content :
    line = line[:-1].split('\t')
    for top in top_inCombinationList :
        top = top[:-1].split('\t')
        if top == line[0]:
            line = line[0]
            toplist.append(line)
    for top in devdata_downlist :
        top = top[:-1].split('\t')
        if top[1] == line[0]:
            line = line[0]
            toplist.append(line)
    for top in testdata_downlist :
        top = top[:-1].split('\t')
        if top[1] == line[0]:
            line = line[0]
            toplist.append(line)
    
np.savetxt("drive/My Drive/TLCN 2/FashionVC/ds224/toplist.dat", np.array(toplist), fmt="%s")

# #############################################################################
```

```
# # lấy danh sách for downlist
downlist = []
with open('drive/My Drive/TLCN 2/FashionVC/dataset/downlist.dat','r') as f:
    content = f.readlines()

for line1 in traindata_list:
  line1 = line1[0:].split('\t')
  downlist.append(line1[1])

for line in content :
    line = line[:-1].split('\t')
    for down in combination_list:
        down = down[:-1].split('\t')
        if down[1] == line[0]:
            line = line[0]
            downlist.append(line)
    for down in devdata_toplist :
        down = down[:-1].split('\t')
        if down[1] == line[0]:
            line = line[0]
            downlist.append(line)
    for down in testdata_toplist :
        down = down[:-1].split('\t')
        if down[1] == line[0]:
            line = line[0]
            downlist.append(line)
    
np.savetxt("drive/My Drive/TLCN 2/FashionVC/ds224/downlist.dat", np.array(downlist), fmt="%s")
```
